```python
import folium
import pandas as pd
```

```python
df = pd.read_csv('電気自動車充電スタンド.csv',encoding='shiftjis')
df.columns
```

```python
df.rename(columns={'徳島県内電気自動車充電スタンド一覧(判明分)': 'ID', 'Unnamed: 1':'スタンド名', 'Unnamed: 2':'郵便番号', 'Unnamed: 3':'住所', '2014/9/30':'緯度', 'Unnamed: 5':'経度', 'Unnamed: 6':'電話番号', 'Unnamed: 7':'URL'},inplace=True)
df=df.drop(0)
#df
```

```python
f_map1 = folium.Map(location=[34.0702, 134.5548], zoom_start=10)
#f_map1
```

```python
for i, row in df.iterrows():
            folium.Marker(
                location=[row['緯度'], row['経度']],
                popup=row['スタンド名'],
                #icon=folium.Icon(color='blue')
            ).add_to(f_map1)
    
f_map1
```
