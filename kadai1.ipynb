```python
df = pd.read_csv("kadai1.csv")
df
```

```python
df["合計"]=df["国語"]+df["数学"]+df["英語"]
df
```

```python
df['氏名'] = df['姓'].str.cat(df['名'], sep=' ')
df.to_csv('kadai1.csv', encoding = encoding)
df
```

```python
df=df.drop(['Unnamed: 0','姓', '名'], axis=1)
df
```

```python
df = df.reindex(columns=['氏名', '性別', '国語', '数学', '英語', '合計'])
df
```

```python
avg=df['合計'].mean()
avg_s = df.groupby("性別").mean()['合計']
```
